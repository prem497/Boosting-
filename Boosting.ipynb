```python
import pandas as pd # Pandas is a library written for the Python programming language for data manipulation and analysis.
import numpy as  np  # numpy is library written for the python programming language for mathematical functions .
```

```python
dataset=pd.read_csv("insurance_pre.csv") #import csv file 
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset=dataset.astype(int)
```

```python
dataset
```

```python
dependent=dataset[['charges']]
dependent
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
independent
                    
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
X_train.values
```

```python
import xgboost as xg
from xgboost import XGBRegressor
from sklearn.svm import SVR
rgr= SVR(C=1.0, epsilon=0.2)
```

```python
rgr.fit(X_train,Y_train)
```

```python
y_pred=rgr.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,y_pred)
```

```python
r_score
```

```python
import pickle
filename="finalized_model_XGboost.sav"
```

```python
pickle.dump(rgr, open(filename,"wb"))
```
